```python
import numpy as np
import matplotlib.pyplot as plt
import MatplotlibSettings
```

```python
# Load data
dl = [np.loadtxt("deg0.dat"), np.loadtxt("deg1.dat"), np.loadtxt("deg2.dat"), np.loadtxt("deg3.dat"), np.loadtxt("deg4.dat"), np.loadtxt("deg5.dat")]
colours = [r"red", r"blue", r"magenta", r"green", r"orange"]
x = dl[0][:,0]

# nodes
nodes = [0.001, 0.00107152, 0.00114815, 0.00123027, 0.00131826, 0.00141254, 0.00151356, 0.00162181]
labels = [r"\textbf{" + str(i) + "}" for i in range(len(nodes))]
```

```python
for i in range(len(dl)):
    plt.xscale("log")
    plt.tick_params(axis = "x", which = "minor", bottom = False, top = False, labelbottom = False)
    plt.xticks(nodes, labels)
    plt.title(r"\textbf{$k =" + str(i) + "$}")
    plt.ylabel(r"$\textbf{$w_\alpha^{(k)}(x)$}$")
    plt.xlabel(r"\textbf{$\alpha$}")
    plt.xlim(0.001, 0.00142)
    plt.ylim(-1, 1.5)
    for j in range(5):
        plt.plot(x, dl[i][:,j+1], color = colours[j], linewidth = 2)

    plt.savefig("deg" + str(i) + ".pdf")
    plt.show()
    plt.close()
```

```python
data = np.loadtxt("coslnx.dat")
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex = "all", gridspec_kw = dict(width_ratios = [1], height_ratios = [3, 1]))
plt.subplots_adjust(wspace = 0, hspace = 0)

ax1.set(ylabel = r"$\cos(\pi\ln(x))+2$")
ax1.set_xlim([0.001, 1])
ax1.set_xscale("log")
ax1.set_ylim([-0.8, 3.5])
ax1.plot(data[:,0], data[:,1], "k--", lw = 1.5)
ax1.plot(data[:,0], data[:,7], color = "green", label = r"$k=0$")
ax1.plot(data[:,0], data[:,2], color = "red", label = r"$k=1$")
ax1.plot(data[:,0], data[:,3], color = "blue", label = r"$k=2$")
ax1.plot(data[:,0], data[:,4], color = "orange", label = r"$k=3$")
ax1.plot(data[:,0], data[:,5], color = "cyan", label = r"$k=4$")
ax1.plot(data[:,0], data[:,6], color = "magenta", label = r"$k=5$")
ax1.legend(ncol=2)

ax2.set(xlabel = r"\textbf{$x$}")
ax2.set(ylabel = r"\textbf{Ratio}")
ax2.set_xlim([0.001, 1])
ax2.set_ylim([0.9, 1.1])
ax2.set_yticks([0.9, 1, 1.1])
ax2.set_yticklabels([r"\textbf{0.9}", r"\textbf{1}", r"\textbf{1.1}"])
ax2.plot(data[:,0], data[:,1]/data[:,1], "k--", lw = 1.5)
ax2.plot(data[:,0], data[:,7]/data[:,1], lw = 1.5, color = "green")
ax2.plot(data[:,0], data[:,2]/data[:,1], lw = 1.5, color = "red")
ax2.plot(data[:,0], data[:,3]/data[:,1], lw = 1.5, color = "blue")
ax2.plot(data[:,0], data[:,4]/data[:,1], lw = 1.5, color = "orange")
ax2.plot(data[:,0], data[:,5]/data[:,1], lw = 1.5, color = "cyan")
ax2.plot(data[:,0], data[:,6]/data[:,1], lw = 1.5, color = "magenta")

plt.savefig("coslnx.pdf")
plt.show()
plt.close()
```

```python
di = np.loadtxt("DerInt.dat")
```

```python
f, (ax1, ax2) = plt.subplots(2, 1, sharex = "all", gridspec_kw = dict(width_ratios = [1], height_ratios = [3, 1]))
plt.subplots_adjust(wspace = 0, hspace = 0)

ax1.set(ylabel = r"$\displaystyle\frac{d}{d}\cos(\pi\ln(x))$")
ax1.set_xlim([0.01, 1])
ax1.set_xscale("log")
ax1.text(0.0025, 1700, r"$\displaystyle\frac{d}{d}\cos(\pi\ln(x))=-\frac{\pi}{x}\sin(\pi\ln(x))$", fontsize=18)
ax1.set_ylim([-900, 2300])
ax1.plot(di[:,0], di[:,4]-di[:,4], "k--", lw = 1.5)
ax1.plot(di[:,0], di[:,4], "k--", lw = 1.5, label = r"\textbf{Analytic}")
ax1.plot(di[:,0], di[:,5], color = "red", label = r"$k=5$")
ax1.plot(di[:,0], [np.pi / x for x in di[:,0]], "b--", lw = 1.5)
ax1.plot(di[:,0], [-np.pi / x for x in di[:,0]], "b--", lw = 1.5)

ax1.legend()

ax2.set(xlabel = r"\textbf{$x$}")
ax2.set(ylabel = r"\textbf{Ratio}")
ax2.set_xlim([0.001, 1])
ax2.set_ylim([0.99, 1.01])
ax2.set_yticks([0.99, 1, 1.01])
ax2.set_yticklabels([r"\textbf{0.99}", r"\textbf{1}", r"\textbf{1.01}"])
ax2.plot(di[:,0], di[:,4]/di[:,5], "k--", lw = 1.5)
ax2.plot(di[:,0], di[:,5]/di[:,5], lw = 1.5, color = "red")

plt.savefig("dcoslnx.pdf")
plt.show()
plt.close()
```
