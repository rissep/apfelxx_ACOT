```python
import numpy as np
import matplotlib.pyplot as plt
import MatplotlibSettings
```

```python
pT = 10
y  = 1
eta = 0
Q  = 91.1876
def phistar(qT):
    qT2 = qT * qT
    Q2  = Q * Q
    pT2 = pT * pT
    csh = np.cosh(y - eta)
    M   = np.sqrt(Q2 + qT2)
    M2  = M * M
    pTl2 = np.sqrt( M2 + pT2 - 2 * pT * M * csh )
    return np.sqrt((Q2 + 2 * pT * ( pT - M * csh + pTl2 )) /
                   (Q2 + 2 * pT * ( pT - M * csh - pTl2 )))* \
                   np.sqrt( 4 * pT * pTl2  / ( Q2 + 4 * pT * pTl2 ) )
```

```python
qTl = np.linspace(0, 100, 1000)
```

```python
plt.xlabel(r"$q_{\rm T}$ \textbf{[GeV]}")
plt.ylabel(r"$\phi^*$")
#plt.text(2.5, 1.400, r"\textbf{$Q = " +str(Q) + "$ GeV}")
#plt.text(2.5, 1.385, r"\textbf{$y = " +str(y) + "$}")
#plt.text(2.5, 1.370, r"\textbf{$p_{T,\ell} = " +str(pT) + "$ GeV}")
#plt.text(2.5, 1.355, r"\textbf{$\eta_{\ell} = " +str(eta) + "$")
plt.xlim(0, 100)
#plt.ylim(1.24, 1.42)
plt.plot(qTl, phistar(qTl), c = "red",  linestyle = "-")
plt.savefig("phistar.pdf")
plt.show()
plt.close()
```
