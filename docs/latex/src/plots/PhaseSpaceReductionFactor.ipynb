```python
import matplotlib.pyplot as pl
import numpy as np

## Define plot parameters
pl.style.use('seaborn-talk')
pl.rc('axes', linewidth=2)
ax = pl.gca()

# Adjust canvas
pl.xlim(0.1, 20)
pl.ylim(0.9, 1.4)
pl.xticks(size=20)
pl.yticks(size=20)

# Axis labels
pl.xlabel(r'$q_T$ [GeV]', fontsize=22, labelpad=-2)
pl.ylabel(r'$\frac{L_{\mu\nu}g^{\mu\nu}}{L_{\mu\nu}g^{\mu\nu}_\perp}$', fontsize=22)

# Title
pl.title(r"$Q = 91$ GeV, $p_{T,\rm min} =$ 20 GeV, $\eta_{\rm max} =$ 2.4", fontsize = 18)

# Line at one
pl.axhline(y=1, xmin=0, xmax=1, linewidth=1, color = 'k', linestyle='--')

# Read data
data = np.loadtxt('PhaseSpaceReductionFactor.dat')

# Plot data
pl.plot(data[:,0], 1 / data[:,1], 'r-', label=r'$y = 0$')
pl.plot(data[:,0], 1 / data[:,2], 'b-', label=r'$y = 1$')
pl.plot(data[:,0], 1 / data[:,3], 'g-', label=r'$y = 2$')

pl.legend(fontsize=22)

pl.savefig('PhaseSpaceReductionFactor.pdf')
```
